````python
# Download and read the data (this may take more than 1 minute)
# .. means go back one
# below this happens two times
# use relative paths
orig_data = pd.read_csv('../../assets/dataset/311-service-requests.csv', parse_dates=['Created Date'], low_memory=False)
````

````python
orig_data.head(8)
````

````python
orig_data.Borough.unique()
````

````python
colors = { 'QUEENS': 'red', "MANHATTAN" : "green", 
          "BROOKLYN" : "orange", "BRONX" : "purple", 
          "STATEN ISLAND" : "yellow", "Unspecified" : "grey" }

m = orig_data.Borough.tolist()
colors = [ colors[borough] for borough in m ]

lats = orig_data.Latitude.tolist()
longs = orig_data.Longitude.tolist()

#markerSymb = {"QUEENS" : "-", "MANHATTAN" : "."}

plt.figure()
for i in range(len(m)):
    plt.scatter(lats[i],longs[i], marker=".", color=colors[i])
plt.show()    
    
# plt.scatter(orig_data['Longitude'], orig_data['Latitude'], marker= ".", color="purple")
````

## Try this Example: 
Graph the number of noise complaints each hour in New York

````python
complaints = orig_data[['Created Date', 'Complaint Type']]
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
noise_complaints.set_index('Created Date').sort_index().resample('H', how=len).plot()
````
